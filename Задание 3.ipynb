```python
from urllib.request import urlopen
from urllib.error import HTTPError, URLError

def check_webpage(url):
    
    try:
        response = urlopen(url)
        print(f"Страница найдена: {url}")
    except HTTPError as e:
        print(f"Ошибка HTTP: {e.code} - Страница не найдена или недоступна.")
    except URLError as e:
        print(f"Ошибка URL: {e.reason} - Проблемы с соединением.")
    else:
        print("Соединение успешно.")


url_to_check = "https://example.com"
check_webpage(url_to_check)
```

```python
import requests

def check_ssl_certificate(url):
    
    try:
        # Делаем GET-запрос к сайту
        response = requests.get(url, timeout=5)
        if response.url.startswith("https://"):
            print(f"Сайт {url} использует SSL-сертификат.")
        else:
            print(f"Сайт {url} не использует SSL-сертификат.")
    except requests.exceptions.SSLError:
        print(f"Ошибка SSL: Сайт {url} не имеет корректного SSL-сертификата.")
    except requests.exceptions.ConnectionError:
        print(f"Ошибка соединения: Не удалось подключиться к {url}.")
    except requests.exceptions.RequestException as e:
        print(f"Произошла ошибка: {e}")
    else:
        print(f"Статус HTTP: {response.status_code}")


url_to_check = "https://example.com"
check_ssl_certificate(url_to_check)
```

```python
import requests

def fetch_website_details(url):
    """
    Получает и выводит информацию о сайте.
    :param url: str - URL веб-сайта
    """
    try:
        
        response = requests.get(url, timeout=5)
        
        
        print("Status Code:", response.status_code)
        print("Headers:", response.headers)
        print("URL:", response.url)
        print("History:", response.history)
        print("Encoding:", response.encoding)
        print("Reason:", response.reason)
        print("Cookies:", response.cookies)
        print("Elapsed Time:", response.elapsed)
        print("Request:", response.request)
        print("Content (first 500 characters):", response.text[:500])  
    except requests.exceptions.RequestException as e:
        print(f"Произошла ошибка: {e}")


url_to_check = "https://python.org"
fetch_website_details(url_to_check)
```

```python
import requests

def fetch_robots_txt(url):
    
    try:
        
        robots_url = f"{url}/robots.txt"
        
        
        response = requests.get(robots_url, timeout=5)
        
        
        if response.status_code == 200:
            print("Содержимое файла robots.txt:")
            print(response.text)
        else:
            print(f"Не удалось получить robots.txt, статус: {response.status_code}")
    except requests.exceptions.RequestException as e:
        print(f"Произошла ошибка: {e}")


url_to_check = "https://en.wikipedia.org"
fetch_robots_txt(url_to_check)
```

```python
import requests
from bs4 import BeautifulSoup

def fetch_h1_from_website(url):
    
    try:
        
        response = requests.get(url, timeout=5)
        
        
        if response.status_code == 200:
            
            soup = BeautifulSoup(response.text, 'html.parser')
            
            
            h1_tag = soup.find('h1')
            if h1_tag:
                print(f"Заголовок <h1>: {h1_tag.text.strip()}")
            else:
                print("Заголовок <h1> не найден на странице.")
        else:
            print(f"Не удалось получить содержимое страницы, статус: {response.status_code}")
    except requests.exceptions.RequestException as e:
        print(f"Произошла ошибка: {e}")


url_to_check = "http://www.example.com"
fetch_h1_from_website(url_to_check)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

def fetch_all_headers(url):
    
    try:
        
        response = urlopen(url)
        
        
        html_content = response.read()
        
        
        soup = BeautifulSoup(html_content, 'html.parser')
        
        
        headers = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
        
        if headers:
            print("Найденные заголовки:")
            for header in headers:
                print(f"{header.name}: {header.text.strip()}")
        else:
            print("Заголовки не найдены на странице.")
    except Exception as e:
        print(f"Произошла ошибка: {e}")


url_to_check = "https://en.wikipedia.org/wiki/Main_Page"
fetch_all_headers(url_to_check)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

def fetch_links(url):
    
    try:
        
        response = urlopen(url)
        html_content = response.read()
        
        
        soup = BeautifulSoup(html_content, 'html.parser')
        
        
        links = soup.find_all('a', href=True)
        
        
        link_list = [link['href'] for link in links]
        
        print("Ссылки на странице:")
        for link in link_list:
            print(link)
        
        return link_list
    except Exception as e:
        print(f"Произошла ошибка: {e}")
        return []


url_to_check = "https://en.wikipedia.org/wiki/Python"
links = fetch_links(url_to_check)
```

```python
import requests
import csv

def count_csv_rows(url):
    
    try:
        
        response = requests.get(url)
        response.raise_for_status()  

       
        lines = response.text.splitlines()
        csv_reader = csv.reader(lines)

        
        row_count = sum(1 for row in csv_reader)

        print(f"Количество строк в CSV-файле: {row_count}")
        return row_count
    except requests.exceptions.RequestException as e:
        print(f"Ошибка при загрузке файла: {e}")
        return 0
    except Exception as e:
        print(f"Произошла ошибка: {e}")
        return 0


csv_url = "http://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/4.5_month.csv"
count_csv_rows(csv_url)
```

```python
import requests
from bs4 import BeautifulSoup
import random

def scrape_allocine_popular_movies():
    
    url = "https://www.allocine.fr/film/meilleurs/"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36",
    }
    
    try:
        
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        
        
        soup = BeautifulSoup(response.text, 'html.parser')
        
        
        movies = []
        
        
        movie_items = soup.find_all('li', class_='mdl')
        print(f"Найдено {len(movie_items)} фильмов на странице.")
        
        for movie in movie_items:
            
            title_tag = movie.find('h2', class_='meta-title')
            title = title_tag.text.strip() if title_tag else "Название не найдено"
            
            
            info_tag = movie.find('div', class_='meta-body-item meta-body-info')
            if info_tag:
                info_text = info_tag.text.strip()
                
                parts = [part.strip() for part in info_text.split('|')]
                genre = parts[0] if len(parts) > 0 else "Жанр не найден"
                year = parts[-1].split()[-1] if len(parts) > 1 else "Год не найден"
            else:
                genre = "Жанр не найден"
                year = "Год не найден"
            
            
            movies.append(f"{title} ({year}) - {genre}")
        
        
        num_movies_to_display = min(10, len(movies))
        random_movies = random.sample(movies, num_movies_to_display)
        
        
        if movies:
            print("Фильмы успешно найдены и извлечены. Вывод 10 случайных фильмов:")
            for movie in random_movies:
                print("--------------------------------------------")
                print(movie)
        else:
            print("Фильмы не найдены.")
            
    except requests.exceptions.RequestException as e:
        print(f"Ошибка при загрузке данных: {e}")
    except Exception as e:
        print(f"Произошла ошибка: {e}")


scrape_allocine_popular_movies()
```
